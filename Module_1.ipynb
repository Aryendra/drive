```python
import pandas as pd
import numpy as np

#ex1.1
dataset=pd.read_csv(r"C:\Users\aryen\OneDrive\Desktop\AMZN.csv")
dataset['Close Price']=dataset['Close Price'].str.replace(",","").astype(float)
dataset['High Price']=dataset['High Price'].str.replace(",","").astype(float)
dataset['Low Price']=dataset['Low Price'].str.replace(",","").astype(float)

#ex1.2
closing_price=dataset['Close Price'].head(90) #data for last 90 days
maximum_price=max(closing_price)
minimum_price=min(closing_price)
mean_of_cp=   closing_price.mean()
print(maximum_price,minimum_price,mean_of_cp)

#ex1.3
dataset['Date']=pd.to_datetime(dataset['Date'])
x=max(dataset['Date'])
y=min(dataset['Date'])
sub=x-y
print("difference:",sub)

#ex1.4
dataset['month']=pd.DatetimeIndex(dataset['Date']).month
dataset['year']=pd.DatetimeIndex(dataset['Date']).year
group_m=dataset.groupby('month')
group_y=dataset.groupby('year')

dataset['Total Traded Quantity']=dataset['Total Traded Quantity'].astype(str).str.replace("M","")
dataset['Total Traded Quantity']=dataset['Total Traded Quantity'].astype(float)
print(dataset['Total Traded Quantity'])
dataset['average price']=(dataset['Close Price']+dataset['High Price']+dataset['Low Price'])/3
print(dataset['average price'])

def vwap(dataset):
    q=dataset['Total Traded Quantity'].values
    p=dataset['average price'].values
    return dataset.assign(vwap=(p*q).cumsum() / q.cumsum())
dataset=dataset.groupby(dataset['month'], group_keys=False).apply(vwap)


#ex1.5
def avg_price(N):
    new_data=dataset['Close Price'].head(N)
    return new_data.mean(axis=0)
print(avg_price(7))

def pro_loss(N):
    return dataset['Close Price'].head(N).pct_change()
print(pro_loss(7))

#ex1.6
dataset['Day_Perc_Change']=dataset['Close Price'].pct_change()
dataset['Day_Perc_Change'].iloc[0]=0

#ex1.7
def upload(s):
    if(s['Day_Perc_Change']>-0.5 and s['Day_Perc_Change']<0.5):
        val='Slight or No change'

        
    else:
        val='Nan'
    return val
dataset['Trend']=dataset.apply(upload, axis=1)
dataset['Total Traded Quantity']=dataset['Total Traded Quantity'].astype(float).replace("M","")


#ex1.8
avg=dataset.groupby('Trend')['Total Traded Quantity'].mean()
med=dataset.groupby('Trend')['Total Traded Quantity'].median()
print("average is :",avg,"\nmedian is :",med)
dataset
```
