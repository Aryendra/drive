```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#ex2.1
dataset=pd.read_csv('week2.csv')
dataset.set_index('Date')
dataset['Date']=pd.to_datetime(dataset['Date'])

plt.plot(dataset['Date'],dataset['Close Price'])
plt.xlabel('date')
plt.ylabel('close price')
plt.show()

#ex2.2
plt.stem(dataset['Day_Perc_Change'],use_line_collection=True)
plt.show()


#ex2.3
plt.stem(dataset['Total Traded Quantity'],use_line_collection=True)

plt.show()
#we can observe from the graph below that there is correlation between the volume and percentage change stock as the volume increases with gain in percentage change and similarly for loss case.


#ex2.4
labels=['Slight or No change']
sizes=[200]
colors=['gold']
explode=[0.1]
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow =True,startangle=140)
plt.axis('equal')
plt.show()
average=dataset[dataset['Trend']=='Slight or No change']['Total Traded Quantity'].mean()
plt.bar(labels,average,color='purple',width=0.1)
plt.show()
median=dataset[dataset['Trend']=='Slight or No change']['Total Traded Quantity'].median()
plt.bar(labels,median,color='maroon',width=0.1)
plt.show()
# in previous dataframe there is only one type of trend i.e. 'Slightly or No change',that's why no other sector in piechart.

#ex2.5
plt.hist(dataset['Day_Perc_Change'],bins=10,density=True)
plt.show()


#ex2.6
import pandas as pd
dataset1=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\MARUTI.csv')
dataset2=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\RELIANCE.csv')
dataset3=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\TITAN.csv')
dataset4=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\TCS.csv')
dataset5=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\BAJFINANCE.csv')


data1={'maruti':dataset1['Close Price'],'reliance':dataset2['Close Price'],'titan':dataset3['Close Price'],'tcs':dataset4['Close Price'],'bajfinance':dataset5['Close Price']}
new1=pd.DataFrame(data1)
print(new1)
data2={'maruti1':new1['maruti'].pct_change(),'reliance1':new1['reliance'].pct_change(),'titan1':new1['titan'].pct_change(),'tcs1':new1['tcs'].pct_change(),'bajfinance1':new1['bajfinance'].pct_change()}
new2=pd.DataFrame(data2)
new2.drop([0],inplace=True)
print(new2)
import pandas as pd
import seaborn as sns
sns.lmplot('maruti1','reliance1',data=new2,fit_reg=False)
sns.lmplot('maruti1','tcs1',data=new2,fit_reg=False)
sns.lmplot('maruti1','titan1',data=new2,fit_reg=False)
sns.lmplot('maruti1','bajfinance1',data=new2,fit_reg=False)

sns.lmplot('reliance1','tcs1',data=new2,fit_reg=False)
sns.lmplot('maruti1','titan1',data=new2,fit_reg=False)
sns.lmplot('maruti1','bajfinance1',data=new2,fit_reg=False)
sns.lmplot('tcs1','titan1',data=new2,fit_reg=False)
sns.lmplot('tcs1','bajfinance1',data=new2,fit_reg=True)

#ex2.7
import matplotlib.pyplot as plt
%matplotlib inline
volatile1=new2['titan1'].rolling(7).std()

#ex2.8
import pandas as pd
nifty=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Nifty50\Nifty50.csv')
nif_pct=nifty['Close'].pct_change()
volatile2=nif_pct.rolling(7).std()
x2=volatile2.head(100)
x1=volatile1.head(100)
plt.plot(x1,label='titan')
plt.plot(x2,label='nifty50')
plt.legend()
plt.show()


#ex2.9
y=dataset3.set_index('Date')
avg_price21=y['Close Price'].head(21).mean()
print(avg_price21)
sma1=y['Close Price'].rolling(21).mean()
sma1.head(10)
avg_price34=y['Close Price'].head(34).mean()
print(avg_price34)
sma2=y['Close Price'].rolling(34).mean()
sma2.head(10)
plt.plot(sma1,label='21day',marker='^',markersize=1)
plt.plot(sma2,label='34day',marker='v',markersize=1)
plt.legend()
plt.show()

#ex2.10
avg_price14=y['Close Price'].rolling(14).mean()
std_14=y['Close Price'].rolling(14).std()
upper_band=avg_price14+(std_14*2)
lower_band=avg_price14-(std_14*2)
plt.plot(avg_price14,label='average mean')
plt.plot(upper_band,label='ub')
plt.plot(lower_band,label='lb')
plt.plot(y['Close Price'],label='average daily')
plt.legend()
plt.show()
y['upper band']=upper_band
y['middle band']=avg_price14
y['lower_band']=lower_band

y.to_csv('week_3.csv')
y
```
