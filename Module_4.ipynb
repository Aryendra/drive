```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#ex4.1
dataset=pd.read_csv('week_3.csv')
dataset
def call(s):
    if(s['Close Price']<s['lower_band']):
        val='Buy'
    elif(s['Close Price']<s['middle band'] and s['Close Price']>s['lower_band']):
        val='Hold Short/Liquidate Buy'
    elif(s['Close Price']>s['middle band'] and s['Close Price']<s['upper band']):
        val='Hold Buy/Liquidate Short'
    else:
        val='Short'
    return val
dataset['Call']=dataset.apply(call, axis=1)

new_dataset=dataset.dropna()


from sklearn.model_selection import train_test_split
X=new_dataset[['Close Price','upper band','middle band','lower_band']]
y=new_dataset['Call']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33)

#using logistic classifier
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,y_train)
predictions=model.predict(X_test)
print(predictions)

from sklearn.metrics import accuracy_score
score=accuracy_score(y_test,predictions)
print("score of lc:",score)

#using random forest classifier
from sklearn.ensemble import RandomForestClassifier
classifier2=RandomForestClassifier()
classifier2.fit(X_train,y_train)
predictions2=classifier2.predict(X_test)

score2=pd.crosstab(y_test,predictions2,rownames=['Actual result'],colnames=['predicted result'])
print("score of rfc",score2)


#using k nearest neighbors classifier
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
predictions3=knn.predict(X_test)

from sklearn import metrics
print('accuracy of k-nearest:',metrics.accuracy_score(y_test,predictions3))


#newstock
new_stock=pd.read_csv(r'C:\Users\aryen\OneDrive\Desktop\Large_Cap\INFY.csv')
avg_price20=new_stock['Close Price'].rolling(20).mean()
std_20=new_stock['Close Price'].rolling(20).std()
upper_band=avg_price20+(std_20*2)
lower_band=avg_price20-(std_20*2)
new_stock['lower_band']=lower_band
new_stock['middle']=avg_price20
new_stock['upper_band']=upper_band
stock_without_null=new_stock.dropna()
def call2(v):
    if(v['Close Price']<v['lower_band']):
        val='Buy'
    elif(v['Close Price']<v['middle'] and v['Close Price']>v['lower_band']):
        val='Hold Short/Liquidate Buy'
    elif(v['Close Price']>v['middle'] and v['Close Price']<v['upper_band']):
        val='Hold Buy/Liquidate Short'
    else:
        val='Short'
    return val
stock_without_null['Call']=stock_without_null.apply(call2, axis=1)

#using logistic classifier for new stock
from sklearn.model_selection import train_test_split
X1=stock_without_null[['Close Price','lower_band','middle','upper_band']]
y1=stock_without_null['Call']
X_train1,X_test1,y_train1,y_test1=train_test_split(X1,y1,test_size=0.4)
from sklearn.linear_model import LogisticRegression
model2=LogisticRegression()
model2.fit(X_train1,y_train1)
predictions1=model2.predict(X_test1)

from sklearn.metrics import accuracy_score
score1=accuracy_score(y_test1,predictions1)
print("score_new:",score1) 



#ex4.2

def pct_change(new_stock):
    

        
    s=(new_stock['Close Price']-new_stock['Open Price'])/new_stock['Close Price']
    return s
new_stock['pct_oc']=new_stock.apply(pct_change, axis=1) 

def pct_change_lh(new_stock):
    
    
    z=(new_stock['High Price']-new_stock['Low Price'])/new_stock['Low Price']
    return z
new_stock['pct_lh']=new_stock.apply(pct_change_lh,axis=1)


new_stock['roll_mean']=new_stock['Close Price'].rolling(5).mean()
new_stock['roll_std']=new_stock['Close Price'].rolling(5).std()

new_stock['Action']=np.where(new_stock['Close Price'].shift(-1)>new_stock['Close Price'],'1','-1')
    
#using random forest classifier
from sklearn.model_selection import train_test_split
X=new_stock.dropna()[['pct_oc','pct_lh','roll_mean','roll_std']]
y=new_stock.dropna()['Action']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.34)
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
model_rf.fit(X_train,y_train)
RandomForestClassifier(bootstrap=True,class_weight=None,criterion='gini',max_depth=None,max_features='auto',max_leaf_nodes=None,min_impurity_split=2,min_weight_fraction_leaf=0.0,n_estimators=10,n_jobs=2,oob_score=False,random_state=0,verbose=0,warm_start=False)
predict=model_rf.predict(X_test)
scoring=pd.crosstab(y_test,predict,rownames=['Actual Result'],colnames=['Predicted Result'])

print(scoring)
from sklearn.metrics import accuracy_score
score4=accuracy_score(y_test,predict)
print("score_new:",score4) 

#graph for net cum returns
cum_returns=np.cumprod(1+new_stock['pct_oc'].values)-1
plt.plot(cum_returns)
plt.show()
new_stock
```
